```python
from darkflow.net.build import TFNet
import cv2
import matplotlib.pyplot as plt
import os

%config InlineBackend.figure_format = 'svg'
```

```python
options = {
    'model': 'cfg/yolo.cfg',
    'load' : 'bin/yolov2.weights',
    'threshold': 0.1,
    'gpu' : 1.0    
}

tfnet = TFNet(options)
```

```python
img = cv2.imread('car1.jpg')
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
result = tfnet.return_predict(img)

result
```

```python
for i in range(len(result)):
    tl = (result[i]['topleft']['x'], result[i]['topleft']['y'])
    br = (result[i]['bottomright']['x'], result[i]['bottomright']['y'])
    label = result[i]['label']

    img = cv2.rectangle(img, tl, br, (255,0,0), 7)
    img = cv2.putText(img,label,br,cv2.FONT_HERSHEY_COMPLEX,1,(0,0,0),2)
        
    plt.figure()
    plt.imshow(img)
    plt.show()

    plt.figure()
    cropped = img[tl[1]:br[1], tl[0]:br[0]]
    cropped = cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB)
    plt.imshow(cropped)
    plt.show()
```

```python
path = 'C:\\AI\darkflow\\darkflow-master\\tryout'
newpath = os.path.join(path, '1.jpg')
print(newpath)
cv2.imwrite(newpath, cropped)
cv2.waitKey(0)

directories = os.listdir(path)
print(directories)
```
