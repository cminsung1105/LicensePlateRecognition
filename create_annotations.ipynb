```python
import os
import matplotlib.pyplot as plt
import cv2
from lxml import etree
import xml.etree.cElementTree as ET
from pathlib import Path
```

```python
def create_annotations(folder):
    directory = os.path.abspath(folder)
    mainpath,savedir = os.path.split(directory)
    savedir = 'annotations'
    savepath = os.path.join(mainpath, savedir)
    
    for file in os.listdir(directory):
        name, ext = os.path.splitext(file)
        if ext == '.txt':        
            path = os.path.join(directory, file)
                                    
            txtfile = open(path, 'r')
            content = txtfile.read().split()
            tl = (int(content[1]), int(content[2]))
            br = (int(content[1])+int(content[3]), 
                  int(content[2])+int(content[4]))
            
            imdir = 'images'
            sourcepath = os.path.join(mainpath,imdir)
            imname = name + '.jpg'
            impath = os.path.join(sourcepath,imname)

            imfolder = os.path.join(imdir, imname)
            
            write_xml(imfolder,impath,imname,name,tl,br,savepath) 
            
            '''
            img = cv2.imread(impath)
            img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
            img = cv2.rectangle(img, tl, br, (255,0,0), 7)
            plt.figure()
            plt.imshow(img)
            plt.show()
            '''
```

```python
def write_xml(folder,impath,imname,name,tl,br,savepath):
    '''
    folder = name of the folder from which the image comes from
    impath = path of the image
    imname = name of the image
    classes = object classes
    tl = top-left corner
    br = bottom-right corner
    savepath = directory to save the xml files
    '''
    
    #Make savepath directory if it does not exist.
    if not os.path.isdir(savepath):
        os.mkdir(savepath)
    
    #Read image
    image = cv2.imread(impath)
    
    #Get Height, Width, Depth
    h,w,d = image.shape
    
    #Highest node in the tree = annotation.
    annotation = ET.Element('annotation')
    
    #Create child node of the tree = folder and name it folder.
    ET.SubElement(annotation,'folder').text = folder
    ET.SubElement(annotation,'filename').text = imname
    ET.SubElement(annotation,'segmented').text = '0'
    
    size = ET.SubElement(annotation,'size')
    ET.SubElement(size,'width').text = str(w)
    ET.SubElement(size,'height').text = str(h)
    ET.SubElement(size,'depth').text = str(d)
    
    ob = ET.SubElement(annotation,'object')
    ET.SubElement(ob,'name').text = 'lp'
    ET.SubElement(ob,'pose').text = 'Unspecified'
    ET.SubElement(ob,'truncated').text = '0'
    ET.SubElement(ob,'difficult').text = '0'
    
    bbox = ET.SubElement(ob,'bndbox')
    ET.SubElement(bbox,'xmin').text = str(tl[0])
    ET.SubElement(bbox,'ymin').text = str(tl[1])
    ET.SubElement(bbox,'xmax').text = str(br[0])
    ET.SubElement(bbox,'ymax').text = str(br[1])
    
    xml_str = ET.tostring(annotation)
    root = etree.fromstring(xml_str)
    xml_str = etree.tostring(root, pretty_print=True)
    
    xmlname = name + '.xml'
    xmlpath = os.path.join(savepath,xmlname)

    with open(xmlpath,'wb') as xmlfile:
        xmlfile.write(xml_str)
        
    xmlfile.close()
```
